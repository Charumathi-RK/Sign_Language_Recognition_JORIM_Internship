```python
from roboflow import Roboflow
```

```python
rf = Roboflow(api_key='YOUR_API_KEY') #ROBOFLOW API KEY
```

```python
#Creating Workspace
workspace =rf.workspace()

workspace.name

workspace.url

workspace.projects()
```

```python
project =workspace.project('sign-language-v8piv')
```

```python
project.version(1)
```

```python
version =project.version(1)
```

```python
model=version.model
```

```python
def predict_image(image_path):
    prediction = model.predict(image_path)
    predictions = prediction.json()
    if 'predictions' in predictions and len(predictions['predictions']) > 0:
        for p in predictions['predictions']:
            class_label = p['class']
            confidence = p['confidence']
            print(f"Class: {class_label}, Confidence: {confidence*100:.2f}%")
    else:
        print("No predictions found!")
    print()
    prediction.plot()
```

```python
img = "/content/Sign-Language-1/test/images/keepsmile40_png.rf.87b6d66215ef6c0fd565ee5be4d76843.jpg"
predict_image(img)
```
